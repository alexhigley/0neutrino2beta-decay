```python
plt.rcParams["figure.figsize"] = [9.00, 5.50]
plt.rcParams["figure.autolayout"] = True

im1 = plt.imread("/Users/alexh/Desktop/integrated_with_II.png")
im2 = plt.imread("/Users/alexh/Desktop/integrated_no_II.png")

fig = plt.figure(figsize=(18,8))

ax1 = fig.add_subplot(121)
im1 = ax1.imshow(im1)
ax1.set_title("a)  With the Image Intensifier", fontsize=15, y=-0.05, pad=-14, fontweight='bold')

plt.axis('off')

ax1 = fig.add_subplot(122)
im1 = ax1.imshow(im2)
ax1.set_title("b)  Without the Image Intensifier", fontsize=15, y=-0.05, pad=-14, fontweight='bold')

plt.axis('off')
plt.show()
```
